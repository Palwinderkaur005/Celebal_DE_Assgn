## **GOLD LAYERS**

### Top industries

```python
from pyspark.sql.functions import sum

top_industry = silver_df.groupBy("Industry") \
    .agg(sum("InvestmentAmount_USD").alias("Total_Investment")) \
    .orderBy("Total_Investment", ascending=False)

display(top_industry)
```

```python
top_industry.write \
    .format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/startup_india_funding/gold_top_industry")
```

### City Ranking

```python
city_rank = silver_df.groupBy("City") \
    .agg(sum("InvestmentAmount_USD").alias("Total_Investment")) \
    .orderBy("Total_Investment", ascending=False)

display(city_rank)
```

```python
city_rank.write \
    .format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/startup_india_funding/gold_city_rank")
```

### Investor Activity

```python
from pyspark.sql.functions import count

investor_activity = silver_df.groupBy("Investors") \
    .agg(count("*").alias("Deals")) \
    .orderBy("Deals", ascending=False)

display(investor_activity)
```

```python
investor_activity.write \
    .format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/startup_india_funding/gold_investor_activity")
```
