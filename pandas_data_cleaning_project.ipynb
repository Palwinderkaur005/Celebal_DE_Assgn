```python
# STEP 1: IMPORT LIBRARIES
import pandas as pd
import numpy as np

# STEP 2: LOAD DATASET
df = pd.read_csv("dresses.csv")
print("Dataset loaded successfully\n")

print("First 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())

print("\nShape of dataset (rows, columns):")
print(df.shape)

print("\nColumn names:")
print(df.columns)

print("\nData types of each column:")
print(df.dtypes)

print("\nBasic info about dataset:")
df.info()

# STEP 3: MISSING VALUES CHECK
print("\nMissing values before cleaning:")
print(df.isnull().sum())

num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing values after cleaning:")
print(df.isnull().sum())

# STEP 4: REMOVE DUPLICATES
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]

print(f"\nNumber of duplicate rows removed: {before - after}")

# STEP 5: CLEANING PRICE COLUMN
df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace(r'[‚āĻ,"]', '', regex=True)
    .str.strip()
)

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# STEP 6: FEATURE ENGINEERING

df['quantity'] = 2
df['total_amount'] = df['final_price'] * df['quantity']

# STEP 7: FILTERING DATA + BASIC INSIGHTS + SAVE CLEANED FILE
print("\nShowing products with rating greater than 4:")
display(df[df['rating'] > 4])

print("\nSome basic insights from data:")
print("Average rating:", df['rating'].mean())
print("Maximum price:", df['final_price'].max())
print("Minimum price:", df['final_price'].min())

df.to_csv("cleaned_dresses.csv", index=False)
print("\nCleaned dataset saved successfully as cleaned_dresses.csv")

# SUMMARY

print("""
FINAL SUMMARY:

- Worked on a dresses dataset containing product details like rating, price, discount, seller info, and categories.

- Loaded the dataset and explored its structure using head(), tail(), shape, columns, data types, and info().

- Checked missing values in columns such as discount, seller_name, what_customers_said, videos, and variations.

- Handled missing values by filling numeric columns with median and categorical columns with mode.

- Removed duplicate rows to ensure clean and accurate data.

- Cleaned the final_price column by removing currency symbols (‚āĻ) and converting it into numeric format.

- Created a new feature called total_amount using final_price √ó quantity.

- Performed filtering to display products with rating greater than 4.

- Generated basic insights such as average rating, maximum price, and minimum price from the dataset.

- Saved the cleaned dataset as 'cleaned_dresses.csv' for further analysis.
""")
```
